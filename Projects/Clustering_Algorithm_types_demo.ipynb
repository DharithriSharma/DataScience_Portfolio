```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.mixture import GaussianMixture

# Generate synthetic data
X, y = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Plot the generated data
plt.scatter(X[:, 0], X[:, 1], s=50, cmap='viridis')
plt.title("Generated Data")
plt.show()
```

```python
# K-Means Clustering
kmeans = KMeans(n_clusters=4)
kmeans_labels = kmeans.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=kmeans_labels, s=50, cmap='viridis')
plt.title("K-Means Clustering")
plt.show()
```

```python
# Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=4)
agg_labels = agg_clustering.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=agg_labels, s=50, cmap='viridis')
plt.title("Agglomerative Clustering")
plt.show()
```

```python
# DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=dbscan_labels, s=50, cmap='viridis')
plt.title("DBSCAN Clustering")
plt.show()
```

```python
# Gaussian Mixture Model Clustering
gmm = GaussianMixture(n_components=4)
gmm_labels = gmm.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=gmm_labels, s=50, cmap='viridis')
plt.title("Gaussian Mixture Model Clustering")
plt.show()
```
